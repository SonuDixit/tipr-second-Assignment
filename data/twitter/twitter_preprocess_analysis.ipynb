```python
with open("twitter.txt","r") as f:
    lines = f.readlines()
```

```python
lines[0]
```

```python
good_lines = [l.rstrip("\n") for l in lines]
```

```python
vocab = {}
for line in good_lines:
    for w in line.split(" "):
        if w in vocab:
            vocab[w] = vocab[w] + 1
        else:
            vocab[w] = 1
```

```python
v = [val for val in vocab.values()]
v.sort()

print(v.count(3))
print(v.count(4))
print(v.count(5))
```

```python
for k,v in vocab.items():
    if v<=3 and (len(k)==1 or len(k) == 2):
        print(k,v)
```

```python
for k,v in vocab.items():
    if v==4 and (len(k)==1 or len(k) == 2):
        print(k,v)
```

```python
for k,v in vocab.items():
    if v==5 and (len(k)==1 or len(k) == 2):
        print(k,v)
```

## removing above kind of non-meaning ful words from vocab

```python
vocab_modified = {}
for k,v in vocab.items():
    if v<=5 and len(k)<=2:
        pass
    else:
        vocab_modified[k] = v
```

```python
len(vocab)-len(vocab_modified)
```

# 153 words removed from vocab

```python
index_dict = {}
i = 0
for k in vocab_modified.keys():
    index_dict[i] = k
    i += 1
```

```python
index_dict
```

```python
index_dict_inv = {v:k for k,v in index_dict.items()}
```

```python
index_dict_inv
```

```python
import numpy as np
```

```python
data = np.zeros((len(lines),len(vocab_modified)))
```

```python
for i in range(len(good_lines)):
    t = good_lines[i].split(" ")
    for j in range(len(t)):
        if t[j] in index_dict_inv:
            data[i][index_dict_inv[t[j]]] += 1
            
```
