```python
import os
from matplotlib.image import imread
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from NN import NN
from helper_fns import prepare_train_test_cat_dog
```

```python
path = os.getcwd()
data_path = os.path.join(os.path.dirname(path),"Data","Cat-Dog")
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
import cv2
```

```python
total_data = []
total_label = []
for label in (["cat","dog"]):
    f = os.listdir(os.path.join(data_path,str(label)))
    for file_name in f:
        a = imread(os.path.join(data_path,str(label),file_name))
        a = rgb2gray(a)
        a = cv2.resize(a,(100,100),interpolation = cv2.INTER_AREA)
        a = a.reshape(a.shape[0] * a.shape[1], )
        total_data.append(a)
        if label == "cat":
            total_label.append(0)
        else:
            total_label.append(1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    total_data, total_label, test_size=0.30, random_state=42,shuffle = True)
```

```python
len(X_train)
```

```python
net_3 = NN(hidden_layer=[30,20,10], activation=["swish","relu","relu"], 
         input_dim=40000,output_dim=2)
```

```python
for _ in range(1):
    for i in range(175):
        data_batch = np.asarray(X_train[i*100: (i+1) * 100])
        label_batch = np.asarray(y_train[i*100: (i+1) * 100])
        y_tr_one_hot = np.zeros((label_batch.shape[0],2))
        for j in range(label_batch.shape[0]):
            y_tr_one_hot[j,label_batch[j]] = 1
        lr = 0.0001
#         if i > 50 :
#             lr = 0.0001
#         if i> 120 :
#             lr = 0.001
        loss = net_3.fit_batch(data_batch,y_tr_one_hot,
                           epochs=1,lr=lr,batch_size=20)
        print(i)
```

## second epoch with _increased_ learning rate

```python
for i in range(175):
    data_batch = np.asarray(X_train[i*100: (i+1) * 100])
    label_batch = np.asarray(y_train[i*100: (i+1) * 100])
    y_tr_one_hot = np.zeros((label_batch.shape[0],2))
    for j in range(label_batch.shape[0]):
        y_tr_one_hot[j,label_batch[j]] = 1
    lr = 0.001
#         if i > 50 :
#             lr = 0.0001
#         if i> 120 :
#             lr = 0.001
    loss = net_3.fit_batch(data_batch,y_tr_one_hot,
                       epochs=1,lr=lr,batch_size=20)
    print(i)
```

```python
pred_vals = net_3.predict(np.asarray(X_test))
# for test_data in X_test:
#     pred_vals.append(net_3.predict(np.asarray(test_data)))
    
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred_vals))
from sklearn.metrics import f1_score
print(f1_score(y_test, pred_vals, average='macro'))
f1_score(y_test, pred_vals, average='micro')
```

```python
net_4 = NN(hidden_layer=[100, 60, 40], activation=["relu","relu","relu"], 
         input_dim=10000,output_dim=2,momentum = 0.8)
```

```python
for _ in range(1):
    for i in range(15):
        data_batch = np.asarray(X_train[i*100: (i+1) * 100])
        label_batch = np.asarray(y_train[i*100: (i+1) * 100])
        y_tr_one_hot = np.zeros((label_batch.shape[0],2))
        for j in range(label_batch.shape[0]):
            y_tr_one_hot[j,label_batch[j]] = 1
        lr = 0.0001
#         if i > 50 :
#             lr = 0.0001
        if i> 3 :
            lr = 0.001
        if i>8:
            lr = 0.01
        loss = net_4.fit_batch(data_batch,y_tr_one_hot,
                           epochs=1,lr=lr,batch_size=20)
        print(i)
```

```python
for e in range(2):
    for i in range(175):
        data_batch = np.asarray(X_train[i*100: (i+1) * 100])
        label_batch = np.asarray(y_train[i*100: (i+1) * 100])
        y_tr_one_hot = np.zeros((label_batch.shape[0],2))
        for j in range(label_batch.shape[0]):
            y_tr_one_hot[j,label_batch[j]] = 1
        lr = 0.01
#         if i > 50 :
#             lr = 0.0001
#         if i> 120 :
#             lr = 0.001
        loss = net_4.fit_batch(data_batch,y_tr_one_hot,
                           epochs=1,lr=lr,batch_size=20)
        print(i,e)
```

```python
pred_vals = net_4.predict(np.asarray(X_test))
# for test_data in X_test:
#     pred_vals.append(net_3.predict(np.asarray(test_data)))
    
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred_vals))
from sklearn.metrics import f1_score
print(f1_score(y_test, pred_vals, average='macro'))
f1_score(y_test, pred_vals, average='micro')
```
